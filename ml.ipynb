```python
import pandas as pd
import random 
# Load the dataset
df = pd.read_csv("E:\ML\HelpingMyMomDecideWhatToCook\indian_food.csv")

# Check basic structure
df.head()
```

```python
# Remove null values
df = df.dropna(subset=['ingredients'])

# Optional: lowercase and remove punctuation
df['ingredients_clean'] = df['ingredients'].str.lower().str.replace('[^a-zA-Z, ]', '', regex=True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Example input from user
user_input = "potato, onion, tomato, rice, turmeric, cumin"

# TF-IDF Vectorization
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df['ingredients_clean'])

# Transform the user's ingredients into the same vector space
user_vector = vectorizer.transform([user_input])

# Compute cosine similarity
similarities = cosine_similarity(user_vector, tfidf_matrix)

# Get top 5 matching recipes
top_indices = similarities.argsort()[0][-5:][::-1]
top_recipes = df.iloc[top_indices][['name', 'ingredients', 'course', 'region']]

print(top_recipes)
```
